```python
fs=open('/content/'+'news_search_result_'+encText+'.csv','r', encoding='utf8')
news_txt = fs.read()
fs.close()
```

```python
okt=Okt()
pos_txt=okt.pos(news_txt)
print(pos_txt)
```

```python
ad =[t[0] for t in pos_txt if t[1]=='Adjective']
n =[t[0] for t in pos_txt if t[1]=='Noun']
```

```python
ad=[]
n=[]

for t in pos_txt:
    if t[1]=="Adjective":
        ad.append(t[0])
    elif t[1]=="Noun":
        n.append(t[0])
```

```python
morph=ad+n
```

```python
import nltk
stop_words=". , 무면허 렌터카 사고 대 운전 명 사망 교통사고 차량 차 세 대들 후 땐 살 건 일보 않는 낸 비대 로 한 차라리 걸 더 쾅 위 불 비 안성 가장 없는 막 몰던 해야 되는".split(" ")
morph_stop=[sw for sw in morph if sw not in stop_words]
```

```python
from collections import Counter
```

```python
morph_cnt=Counter(morph_stop)
morph_txt=morph_cnt.most_common()
```
