```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import requests
```

```python
keyword='무면허 렌터카 사고'

driver = webdriver.Chrome('C:/pydata/chromedriver.exe')
driver.get('https://news.google.com/?hl=ko&gl=KR&ceid=KR%3Ako')
driver.implicitly_wait(5)
```

```python
search = driver.find_element_by_xpath(
    '//*[@id="gb"]/div[2]/div[2]/div/form/div[1]/div/div/div/div/div[1]/input[2]')

search.send_keys(keyword)
search.send_keys(Keys.ENTER)
driver.implicitly_wait(30)
```

```python
url = driver.current_url
rep = requests.get(url)
soup = bs(resp.text, 'lxml')
```

```python
titles = []
links = []

for link in soup.select('h3 > a'):
    href = 'https://news.google.com' + link.get('href')[1:]
    title = link.string
    titles.append(title)
    links.append(href)
```

```python
data = {'title': titles, 'link': links}
data_frame = pd.DataFrame(data, columns=['title', 'link'])
data_frame.to_csv('./' + keyword + '.csv')
```

```python
from konlpy.tag import Okt
from nltk import Text
from wordcloud import WordCloud
```

```python
import csv
data_csv = open("C:/Users/cyd94/Desktop/pyCODE/무면허 렌터카 사고.csv", "r",encoding='UTF8')
texts=data_csv.read()
data_csv.close()
```

```python
t=Okt()
pos_txt=t.pos(texts)
nouns_txt=t.nouns(texts)
print(nouns_txt)
```

```python
ad =[t[0] for t in pos_txt if t[1]=='Adjective'
n =[t[0] for t in pos_txt if t[1]=='Noun']
v =[t[0] for t in pos_txt if t[1]=='Verb']
```

```python
ad=[]
n=[]
v=[]

for t in pos_txt:
    if t[1]=="Adjective":
        ad.append(t[0])
    elif t[1]=="Noun":
        n.append(t[0])
    elif t[1]=="Verb":
        v.append(t[0])
```

```python
morph=ad+n+v
```

```python
import nltk
stop_words=". , 무면허 렌터카 사고 대 운전 명 사망 교통사고 차량 차 세 대들 후 땐 살 건 일보 않는 낸 비대 로 한 차라리 걸 더 쾅 위 불 비 안성 가장 없는 막 몰던 해야 되는".split(" ")
morph_stop=[sw for sw in morph if sw not in stop_words]
```

```python
from collections import Counter
```

```python
morph_cnt=Counter(morph_stop)
morph_txt=morph_cnt.most_common()
```

```python
morph_txt=dict(morph_txt)
print(morph_txt)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(font_path="c:/Windows/Fonts/malgun.ttf",
                      background_color="white", width=1000, height=1000,
                      max_words=100, max_font_size=300)

wc = wordcloud.generate_from_frequencies(morph_txt)
plt.figure(figsize=(8,15))
plt.imshow(wc)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
